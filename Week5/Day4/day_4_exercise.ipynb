```python
Why is data visualization important?
Because it's much easier to understand data when you can see it as a chart or graph instead of staring at rows of numbers. It helps you spot patterns and trends quickly.

What is a line graph for?
A line graph shows how something changes over time. For example, tracking sales every month — you can easily see if they're going up or down.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
days = [1, 2, 3, 4, 5, 6, 7]
temps = [72, 74, 76, 80, 82, 78, 75]
plt.plot(days, temps)
plt.title("Temperature Over Time")
plt.xlabel("days")
plt.ylabel("temps")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

months = ["Jan", "Feb", "Mar", "Apr", "May"]
monthly_sales = [5000, 5500, 6200, 7000, 7500]

plt.bar(months, monthly_sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
```

```python

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = pd.read_csv("smh.csv")

sns.barplot(data=df, x='Choose your gender', y='Do you have Anxiety?', hue='Choose your gender', palette='Set2', legend=False)
plt.title("Anxiety levels in genders")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("smh.csv")

df['Do you have Panic attack?'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

sns.scatterplot(data=df, x='Age', y='Do you have Panic attack?')
plt.title("Relationship Between Age and Panic Attacks")
plt.xlabel("Age")
plt.ylabel("Panic Attack (1=Yes, 0=No)")
plt.show()
```
